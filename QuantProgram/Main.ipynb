```python
close = stocks.loc[:,'Close'].copy(deep=True)
close
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
close.plot(figsize=(15, 8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
close.AAPL.div(close.AAPL[0])
```

```python
norm_close = close.div(close.iloc[0,:]).mul(100)
```

```python
norm_close.plot(figsize=(15, 8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
aapl=close.AAPL.copy(deep = True).to_frame()
aapl
```

```python
aapl['lag1'] = aapl.shift(periods=1)
aapl
```

```python
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl['% Change'] = aapl.Diff.div(aapl.lag1).mul(100)
aapl
```

```python
aapl['Diff2']=aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl['% Change2']=aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```
